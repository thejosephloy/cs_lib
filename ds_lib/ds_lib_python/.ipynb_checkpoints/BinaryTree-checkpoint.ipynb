```python
# Simple B Tree - Does not maintain nlg(n) tree height but attempts to store objects in order from min to max
class Node:
    def __init__(self, x = None, left = None, right = None):
        self.value = x
        self.left = left
        self.right = right
    
class BTree:
    def __init__(self, x = None):
        self.root = x
        
    def traverse(self, node):
        if (node.left == None and node.right == None):
            print(node.value)
            return
        else:
            if (node.left != None):
                self.traverse(node.left)
            print(node.value)
            if (node.right != None):
                self.traverse(node.right)
            return
        
    def find(self, node, key):
        if (node.value == key):
            print("Key found")
            return
        else:
            if (node.left != None):
                self.find(node.left, key)
            if (node.right != None):
                self.find(node.right, key)
            return
        
    def insert(self, node, key):
        if (node.value == key):
            print("Key is already stored in tree")
            return
        elif (node.right == None and node.left == None):
            if (key < node.value):
                node.left = Node(key)
                return
            else:
                node.right = Node(key)
                return
        else:
            if (key < node.value):
                if (node.left == None):
                    node.left = Node(key)
                else:
                    self.insert(node.left, key)
            elif (key > node.value):
                if (node.right == None):
                    node.right = Node(key)
                else:
                    self.insert(node.right, key)
            else:
                return
        return
        
    def delete(self, node, key):
        if (key == self.root):
            return
        return
        
```

```python
x2 = Node(2)
bt1 = BTree(x2)
bt1.insert(bt1.root, 4)
bt1.insert(bt1.root, 0)
bt1.insert(bt1.root, 7)
bt1.insert(bt1.root, 6)
bt1.insert(bt1.root, 9)
bt1.traverse(bt1.root)
bt1.find(bt1.root, 7)
```
