```python
class Node:
    def __init__(self, x_val, next=None):
        self.value = x_val
        self.next = next

class LinkedList:

    def __init__(self, x=None):
        self.HEAD = x
        
        if (x == None):
            self.len = 0
        else:
            self.len = 1
        
        self.TAIL = x
        
    def traverse(self):
        ll_iter = self.HEAD
        for idx in range(self.len):
            print('ll[' + str(idx) + '] is ' + str(ll_iter.value))
            ll_iter = ll_iter.next
    
    def get_at(self, i):
        ll_iter = self.HEAD
        for idx in range(i - 1):
            ll_iter = ll_iter.next
        print(ll_iter.value)
    
    def insert_first(self, x):
        tmp = self.HEAD
        self.HEAD = x
        x.next = tmp
        self.len = self.len + 1
       
    def delete_first(self):
        self.HEAD = self.HEAD.next
        self.len = self.len - 1
    
    def insert_last(self, x):
        self.TAIL.next = x
        self.TAIL = x
        self.TAIL.next = None
        self.len = self.len + 1
        
    
    def delete_last(self):
        ll_iter = self.HEAD
        for idx in range(self.len - 1):
            print('ll[' + str(idx) + '] is ' + str(ll_iter.value))
            ll_iter = ll_iter.next
        ll_iter.next = None
        self.len = self.len - 1
        
    def insert_at(self, i, x):
        ll_iter = self.HEAD
        for idx in range(i - 1):
            ll_iter = ll_iter.next
        tmp = ll_iter.next
        ll_iter.next = x
        x.next = tmp
        self.len = self.len + 1

    def delete_at(self, i):
        ll_iter = self.HEAD
        for idx in range(i - 1):
            ll_iter = ll_iter.next
        ll_iter.next = ll_iter.next.next
        self.len = self.len - 1
```

```python
x1 = Node(1)
x1a = Node(1)
x2 = Node(2)
x3 = Node(3)
ll1 = LinkedList(x3)

ll1.insert_first(x2)
ll1.insert_first(x1)
ll1.insert_at(2, x1a)
ll1.traverse()
ll1.delete_at(2)
ll1.traverse()
```
