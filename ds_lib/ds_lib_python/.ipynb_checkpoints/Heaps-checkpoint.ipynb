```python
# Interace described by CLRS
import math
class BHeap:
    def __init__(self, x = []):
        self.A = x
        self.length = len(self.A)
        self.heap_size = len(self.A)
        
    def parent(self, i):
        return math.floor(i / 2) 
        
    def left(self, i):
        return (2 * i) + 1
    
    def right(self, i):
        return (2 * i) + 2
        
    def max_heapify(self, i):
        l = self.left(i)
        r = self.right(i)
        if (l <= self.heap_size - 1 and self.A[l] > self.A[i]):
            largest = l
        else:
            largest = i
        if (r <= self.heap_size - 1 and self.A[r] > self.A[largest]):
            largest = r
        if (largest != i):
            tmp = self.A[i]
            self.A[i] = self.A[largest]
            self.A[largest] = tmp
            self.max_heapify(largest)
         
    def build_max_heap(self):
        self.heap_size = self.length
        # for i from floor(length / 2) downto 1
        for i in range(math.floor(self.length / 2), -1, -1):
            self.max_heapify(i)
        
    def heapsort(self):
        self.build_max_heap()
        for i in range(self.length - 1, 0, -1):
            tmp = self.A[0]
            self.A[0] = self.A[i]
            self.A[i] = tmp
            self.heap_size = self.heap_size - 1
            self.max_heapify(0)
       

    
MP1 = BHeap([1, 5, 65, 43, 21, 67, 87, 65])
print(MP1.A)
MP1.build_max_heap()
print(MP1.A)
MP1.heapsort()
print(MP1.A)
```

```python
## Huffman Encode
# Priority queue using a binary min heap 
INTMAX = 2147483647
import math
class minPriorityQueue:
    def __init__(self, x = []):
        self.A = x
        self.length = len(self.A)
        self.heap_size = len(self.A)
        
    def parent(self, i):
        return math.floor(i / 2) 
        
    def left(self, i):
        return (2 * i) + 1
    
    def right(self, i):
        return (2 * i) + 2
        
    def min_heapify(self, i):
        l = self.left(i)
        r = self.right(i)
        if (l <= self.heap_size - 1 and self.A[l] < self.A[i]):
            smallest = l
        else:
            smallest = i
        if (r <= self.heap_size - 1 and self.A[r] < self.A[smallest]):
            smallest = r
        if (smallest != i):
            tmp = self.A[i]
            self.A[i] = self.A[smallest]
            self.A[smallest] = tmp
            self.min_heapify(smallest)
         
    def build_min_heap(self):
        self.heap_size = self.length
        # for i from floor(length / 2) downto 1
        for i in range(math.floor(self.length / 2), -1, -1):
            self.min_heapify(i)
            
    def heap_extract_min(self, x):
        if (self.heap_size < 1):
            raise Exception("heap underflow")
        minVal = self.A[0]
        self.A[0] = self.A[self.heap_size - 1]
        self.heap_size = self.heap_size - 1
        self.min_heapify(0)
        return minVal
        
    def heap_minimum(self):
        return self.A[0]
        
    
    def heap_decrease_key(self, i, key):
        if (key > self.A[i]):
            raise Exception("new key is bigger than current key")
        self.A[i] = key
        while (i > 0 and self.A[self.parent(i)] > self.A[i]):
            tmp = self.A[i]
            self.A[i] = self.A[self.parent(i)]
            self.A[self.parent(i)] = tmp
            i = self.parent(i)
        
    def min_heap_insert(self, key):
        self.heap_size = self.heap_size + 1
        self.A.append(INTMAX)
        self.heap_decrease_key(self.heap_size - 1, key)
```

```python
MP1 = minPriorityQueue([1, 5, 65, 43, 21, 67, 87, 65])
print(MP1.A)
MP1.build_min_heap()
print(MP1.A)
```

```python
# Priority queue using a binary max heap 
import math
INTMAX = 2147483647
class maxPriorityQueue:
    def __init__(self, x = []):
        self.A = x
        self.length = len(self.A)
        self.heap_size = len(self.A)
        
    def parent(self, i):
        return math.floor(i / 2) 
        
    def left(self, i):
        return (2 * i) + 1
    
    def right(self, i):
        return (2 * i) + 2
        
    def max_heapify(self, i):
        l = self.left(i)
        r = self.right(i)
        if (l <= self.heap_size - 1 and self.A[l] > self.A[i]):
            largest = l
        else:
            largest = i
        if (r <= self.heap_size - 1 and self.A[r] > self.A[largest]):
            largest = r
        if (largest != i):
            tmp = self.A[i]
            self.A[i] = self.A[largest]
            self.A[largest] = tmp
            self.max_heapify(largest)
         
    def build_max_heap(self):
        self.heap_size = self.length
        # for i from floor(length / 2) downto 1
        for i in range(math.floor(self.length / 2), -1, -1):
            self.max_heapify(i)
            
    def heap_extract_max(self, x):
        if (self.heap_size < 1):
            raise Exception("heap underflow")
        maxVal = self.A[0]
        self.A[0] = self.A[self.heap_size - 1]
        self.heap_size = self.heap_size - 1
        self.max_heapify(0)
        return maxVal
        
    def heap_maximum(self):
        return self.A[0]
        
    
    def heap_increase_key(self, i, key):
        if (key < self.A[i]):
            raise Exception("new key is smaller than current key")
        self.A[i] = key
        while (i > 0 and self.A[self.parent(i)] < self.A[i]):
            tmp = self.A[i]
            self.A[i] = self.A[self.parent(i)]
            self.A[self.parent(i)] = tmp
            i = self.parent(i)
        
    def max_heap_insert(self, key):
        self.heap_size = self.heap_size + 1
        self.A.append(-INTMAX)
        self.heap_increase_key(self.heap_size - 1, key)
```

```python
MP1 = maxPriorityQueue([1, 5, 65, 43, 21, 67, 87, 65])
print(MP1.A)
MP1.build_max_heap()
print(MP1.A)
```
