```python
# Interace described by CLRS
class Node:
    def __init__(self, x = None, left = None, right = None, parent = None):
        self.value = x
        self.left = left
        self.right = right
        self.parent = parent
    
class BSTree:
    def __init__(self, x = None):
        self.root = x
        
    def inorder_tree_walk(self, x):
        if (x != None):
            self.inorder_tree_walk(x.left)
            print(x.value)
            self.inorder_tree_walk(x.right)
    
    
    def tree_search(self, x, key):
        if (x == None or key == x.value):
            return x
        if (key < x.value):
            return self.tree_search(x.left, key)
        else:
            return self.tree_search(x.right, key)
        
       
    def iterative_tree_search(self, x, key):
        while(x.value != None and key != x.value):
            if (key < x.value):
                x = x.left
            else:
                x = x.right
        return x
         
    def tree_minimum(self, x):
        while (x.left != None):
            x = x.left
        return x
        
    def tree_maximum(self, x):
        while (x.right != None):
            x = x.right
        return x
        
    def tree_succesor(self, x):
        if (x.right != None):
            return self.tree_minimum(x.right)
        y = x.parent
        while (y != None and x == y.right):
            x = y
            y = y.parent
        return y
        
        
       
    def tree_predecesor(self, x):
        if (x.left != None):
            return self.tree_maximum(x.left)
        y = x.parent
        while (y != None and x == y.left):
            x = y
            y = y.parent
        return y
     
 
    def tree_insert(self, z):
        y = None
        x = self.root
        print(z)
        print(z.value)
        while (x != None):
            y = x
            if (z.value < x.value):
                x = x.left
            else:
                x = x.right
        z.parent = y
        if (y == None):
            self.root = z
        elif (z.value < y.value):
            y.left = z
        else:
            y.right = z
        
    
    def transplant(self, u, v):
        if (u.parent == None):
            self.root == v
        elif (u == u.parent.left):
            u.parent.left = v
        else:
            u.parent.right = v
        if (v != None):
            v.parent = u.parent
        
        
    def tree_delete(self, z):
        if (z.left == None):
            self.transplant(z, z.right)
        elif (z.right == None):
            self.transplant(z, z.left)
        else:
            y = self.tree_minimum(z.right)
            if (y.parent != z):
                self.transplant(y, y.right)
                y.right = z.right
                y.right.parent = y
            self.transplant(z, y)
            y.left = z.left
            y.left.parent = y


    
x6 = Node(6)
x5a = Node(5)
x5b = Node(5)
x7 = Node(7)
x8 = Node(8)
x2 = Node(2)
x6.left = x5a
x6.right = x7
x5a.left = x2
x5a.right = x5b
x7.right = x8
x5a.parent = x6
x2.parent = x5a
x5b.parent = x5a
x7.parent = x6
x8.parent = x7

bstree = BSTree(x6)
bstree.inorder_tree_walk(bstree.root)
print("\n")
x9 = Node(x9)
bstree.tree_insert(x9)
bstree.inorder_tree_walk(bstree.root)
print("\n")
bstree.tree_delete(x5b)
bstree.inorder_tree_walk(bstree.root)
```

```python
# Interace described by CLRS
import math
class BHeap:
    def __init__(self, x = []):
        self.A = x
        self.length = len(self.A)
        self.heap_size = len(self.A)
        
    def parent(self, i):
        return math.floor(i / 2) 
        
    def left(self, i):
        return (2 * i) + 1
    
    def right(self, i):
        return (2 * i) + 2
        
    def max_heapify(self, i):
        l = self.left(i)
        r = self.right(i)
        if (l <= self.heap_size - 1 and self.A[l] > self.A[i]):
            largest = l
        else:
            largest = i
        if (r <= self.heap_size - 1 and self.A[r] > self.A[largest]):
            largest = r
        if (largest != i):
            tmp = self.A[i]
            self.A[i] = self.A[largest]
            self.A[largest] = tmp
            self.max_heapify(largest)
         
    def build_max_heap(self):
        self.heap_size = self.length
        # for i from floor(length / 2) downto 1
        for i in range(math.floor(self.length / 2), -1, -1):
            self.max_heapify(i)
        
    def heapsort(self):
        self.build_max_heap()
        for i in range(self.length - 1, 0, -1):
            tmp = self.A[0]
            self.A[0] = self.A[i]
            self.A[i] = tmp
            self.heap_size = self.heap_size - 1
            self.max_heapify(0)
       
```
